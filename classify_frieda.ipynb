```python
#path = Path("/content/gdrive/MyDrive/Colab Notebooks/frieda_classifier.pkl")
path = Path("./frieda_classifier.pkl")
learner_inf = load_learner(path)
```

```python
btn_pred = widgets.Button(description="Classify")
out_pl = widgets.Output()
lbl_pred = widgets.Label()

def on_click_classify(change):
  img = PILImage.create(btn_upload.data[-1])
  out_pl.clear_output()
  with out_pl: display(img.to_thumb(128, 128))
  pred, pred_idx, probs = learner_inf.predict(img)
  val = f"Prediction '{pred}' idx {pred_idx} probs {probs}"
  lbl_pred.value = val

btn_pred.on_click(on_click_classify)
btn_upload = widgets.FileUpload()
lbl = widgets.Label("Select your image")

VBox([lbl, btn_upload, btn_pred, out_pl, widgets.Label("Categories are: " + str(learner_inf.dls.vocab)), lbl_pred])
```
